## 1. (수학_48) 다음 행렬을 고유값과 고유벡터로 대각화하세요. (15점)
$$\mathbf{B} = \begin{bmatrix} 4 & 3 \\ 2 & 1 \end{bmatrix}$$

- 힌트: 고유값(w)과 고유벡터행렬(V)은 np.linalg.eig(), 고유값행렬은 np.diag(), 고유벡터행렬의 역행렬(V_inv)은 np.linalg.inv(), 내적은 @로 구하세요.

\
- 주어진 행렬을 코드로 작성한다. (5점)
- 고유값행렬과 고유벡터행렬, 고유벡터행렬의 역행렬을 구한다. (5점)
- 내적을 통해 결과를 확인한다. (5점)

````
# 여기에 코드를 작성해주세요.
import numpy as np
B = np.array([[4,3], [2,1]])
w, V = np.linalg.eig(B)
V_inv = np.linalg.inv(V)
diagB = np.diag(w)

print(f"고유값행렬:\n {diagB}")
print(f"고유벡터행렬:\n {V}")
print(f"고유벡터행렬의 역행렬:\n {V_inv}")
print(f"내적으로 결과 확인: \n {V @ diagB @ V_inv}")
````

## 5. (수학_87) 다음 정적분의 값을 부정적분과 수치적분 두 가지 방법으로 구하세요. (20점)
$$ \int_0^4 (x^2 + 3exp(x)) dx  $$

- 문제 1: SymPy를 사용하여 부정적분 방법으로 구하세요. (10점)
  - 힌트: 심볼 변수 정의는 symbols()로 구하세요.
- 문제 2: 수치적분 방법으로 구하세요. (10점)
  - 힌트: 수치적분은 sp.integrate.quad()로 구하세요.

````
# 여기에 코드를 작성해주세요.
import scipy as sp
import sympy
sympy.init_printing(use_latex='mathjax')
x  = sympy.symbols('x')
f = x**2 + 3*sympy.exp(x)

#부정 적분
F = sympy.integrate(f)
defi = (F.subs(x,4) - F.subs(x,0)).evalf()

#수치 적분 
def f1(x):
  return x**2 + 3*sympy.exp(x)
indef = sp.integrate.quad(f1, 0, 4)[0] 

print(f"부정 적분 결과:\n {defi}")
print(f"수치 적분 결과:\n {indef}")
````

## 6. (수학_97, 수학_98) 다음 함수를 최적화하세요. (20점)

$$ f_2(x, y) = (1 - x)^2 + 100(y - x^2)^2 $$


- 문제: 그레디언트벡터 함수를 구현하여 jac 인수로 주는 방법으로 계산 속도를 향상시키세요.
  - 조건: 초깃값 x0 = (2, 0.3)
  - 힌트: 최적화는 sp.optimize.minimize()로 구하세요.

````
# 여기에 코드를 작성해주세요.
import scipy as sp
import numpy as np
import scipy.optimize
def f1(x):
  return (1-x[0])**2 + 100*(x[1]-x[0]**2)**2

def fj(x):
  return np.array((-400*x[0]*(-x[0]**2 + x[1])+2*x[0]-2, -200*x[0]**2+200*x[1]))

x0 = (2, 0.3)
result = sp.optimize.minimize(f1, x0, jac = fj)
print(result)
````
