## 1. (ìˆ˜í•™_101) SciPyë¥¼ ì‚¬ìš©í•˜ì—¬ ì•„ë˜ ë¬¸ì œë¥¼ ê³„ì‚°í•˜ì„¸ìš”. (20ì )

ì œí•œì¡°ê±´ì´ $ x_1 + x_2 = 1 $ ì¼ ë•Œ,
\
ëª©ì  í•¨ìˆ˜ $ f(x) = - log x_1 - log x_2 $ ë¥¼ ìµœì†Œí™”í•˜ëŠ” $ x_1, x_2 $ ê°’ì„ ë¼ê·¸ë‘ì£¼ ìŠ¹ìˆ˜ë²•ìœ¼ë¡œ ê³„ì‚°í•˜ì„¸ìš”($ x_1, x_2 > 0 $).

\
- íŒíŠ¸: ìµœì í™”ëŠ” sp.optimize.fmin_slsqp()ë¡œ êµ¬í•˜ì„¸ìš”.

\
- ëª©ì í•¨ìˆ˜ë¥¼ í•¨ìˆ˜ë¡œ ì‘ì„±í•œë‹¤. (5ì )
- ì œí•œì¡°ê±´ì„ í•¨ìˆ˜ë¡œ ì‘ì„±í•œë‹¤. (5ì )
- fmin_slsqp() ëª…ë ¹ì„ ì‚¬ìš©í•˜ì—¬ ë‹µì„ ê³„ì‚°í•œë‹¤. (10ì )

````python
# ì—¬ê¸°ì— ì½”ë“œë¥¼ ì‘ì„±í•´ì£¼ì„¸ìš”.
import numpy as np
import scipy as sp
import scipy.optimize

def f(x):
  return -np.log(x[0]) -np.log(x[1])

def constraint(x):
  return x[0] + x[1] - 1

sp.optimize.fmin_slsqp(f, np.array([1, 1]), eqcons=[constraint])
````

````python
# ì—¬ê¸°ì— ì½”ë“œë¥¼ ì‘ì„±í•´ì£¼ì„¸ìš”.
import numpy as np
def P(a, b):
  if a > b:
    raise ValueError("a must be less than b or equal to b")
  a = np.maximum(a, 0)
  b = np.minimum(b, 360)

  if b < 180:
    return (2/3) * ((b-a)/180)
  else:
    if a < 180:
      return (2/3)*((180-a)/180) + (1/3)*((b - 180)/180)
    return (1/3) * (((b-a)/180))

P(0, 270)
````
