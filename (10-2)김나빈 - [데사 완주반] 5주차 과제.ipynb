## 1. (수학_101) SciPy를 사용하여 아래 문제를 계산하세요. (20점)

제한조건이 $ x_1 + x_2 = 1 $ 일 때,
\
목적 함수 $ f(x) = - log x_1 - log x_2 $ 를 최소화하는 $ x_1, x_2 $ 값을 라그랑주 승수법으로 계산하세요($ x_1, x_2 > 0 $).

\
- 힌트: 최적화는 sp.optimize.fmin_slsqp()로 구하세요.

\
- 목적함수를 함수로 작성한다. (5점)
- 제한조건을 함수로 작성한다. (5점)
- fmin_slsqp() 명령을 사용하여 답을 계산한다. (10점)

````python
# 여기에 코드를 작성해주세요.
import numpy as np
import scipy as sp
import scipy.optimize

def f(x):
  return -np.log(x[0]) -np.log(x[1])

def constraint(x):
  return x[0] + x[1] - 1

sp.optimize.fmin_slsqp(f, np.array([1, 1]), eqcons=[constraint])
````
