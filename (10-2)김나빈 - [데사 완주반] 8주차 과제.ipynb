## 3. (회귀분석_38) 레버리지가 높은 데이터를 포함한 100개의 가상 데이터의 선형회귀를 한 후, 그 결과를 이용하여 레버리지가 큰 아웃라이어를 그래프로 확인하세요. (20점)

\
- 힌트: 데이터 생성은 make_regression(), 그래프는 sm.graphics.plot_leverage_resid2()로 그리세요.

\
- 레버리지가 높은 데이터를 포함한 가상의 데이터를 생성한다. (5점)
- 생성한 가상 데이터를 선형회귀한다. (5점)
- 그래프를 그려 레버리지가 큰 아웃라이어를 확인한다. (10점)

```python
# 여기에 코드를 작성해주세요
from sklearn.datasets import make_regression
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# 100개의 데이터 생성
X0, y, coef = make_regression(n_samples=100, n_features=1, noise=20,
                              coef=True, random_state=1)

# 레버리지가 높은 가상의 데이터를 추가
data_100 = (4, 300)
data_101 = (3, 150)
X0 = np.vstack([X0, np.array([data_100[:1], data_101[:1]])])
X = sm.add_constant(X0)
y = np.hstack([y, [data_100[1], data_101[1]]])

model = sm.OLS(pd.DataFrame(y), pd.DataFrame(X))
result = model.fit()

sm.graphics.plot_leverage_resid2(result)
plt.title("Leverage vs Normalized residuals squared")
plt.xlabel("Normalized residuals**2")
plt.ylabel("Lverage")
plt.show()

```

- 조건 1: 교차검증 5번 반복하고, 평가 지표는 결정 계수로 한다(random_state=4). (5점)
- 조건 2: 아래의 wrapper class와 cross_val_score 명령을 사용하세요. (5점)

    # 아래 코드를 먼저 실행하세요.

    import statsmodels.api as sm

    from sklearn.base import BaseEstimator, RegressorMixin

\
- 교차검증을 반복하는 코드를 작성한다. (10점)

```python
# 여기에 코드를 작성해주세요
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.base import BaseEstimator, RegressorMixin
from sklearn.datasets import load_boston
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
import numpy as np
import pandas as pd

class StatsmodelsOLS(BaseEstimator, RegressorMixin):
  def __init__(self, formula):
    self.formula = formula
    self.model = None
    self.data = None
    self.result = None

  def fit(self, dfX, dfy):
    self.data = pd.concat([dfX, dfy], axis=1)
    self.model = smf.ols(self.formula, data=self.data)
    self.result = self.model.fit()

  def predict(self, new_data):
    return self.result.predict(new_data)

boston = load_boston()
dfX = pd.DataFrame(boston.data, columns=boston.feature_names)
dfy = pd.DataFrame(boston.target, columns=["MEDV"])
df = pd.concat([dfX, dfy], axis=1)

model = StatsmodelsOLS("MEDV ~ " + "+".join(boston.feature_names))
cv = KFold(5, shuffle=True, random_state=4)
cross_val_score(model, dfX, dfy, scoring="r2", cv=cv)

#array([0.72634515, 0.7335056 , 0.74050379, 0.77033146, 0.66292892])
```
