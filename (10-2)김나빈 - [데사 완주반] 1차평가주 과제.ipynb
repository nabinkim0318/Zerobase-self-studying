```
# 생성된 점들을 2차원 삼각형 위에서 볼 수 있도록 그려주는 함수

def plot_triangle(X, kind):
  n1 = np.array([1, 0, 0])
  n2 = np.array([0, 1, 0])
  n3 = np.array([0, 0, 1])
  n12 = (n1 + n2) / 2
  m1 = np.array([1, -1, 0])
  m2 = n3 - n12

  m2 = m2 / np.linalg.norm(m2)

  X1 = (X - n12).dot(m1)

```

\
- 그래프를 그린다. (5점)

````python
# 여기에 코드를 작성해주세요.
import numpy as np
import seaborn as sns
import scipy as sp

def plot_triangle(X, kind):
    n1 = np.array([1, 0, 0])
    n2 = np.array([0, 1, 0])
    n3 = np.array([0, 0, 1])
    n12 = (n1 + n2) / 2
    m1 = np.array([1, -1, 0])
    m2 = n3 - n12
    m1 = m1 / np.linalg.norm(m1)
    m2 = m2 / np.linalg.norm(m2)

    X1 = (X - n12).dot(m1)
    X2 = (X - n12).dot(m2)

    sns.jointplot(X1, X2, kind=kind, xlim=(-0.8, 0.8), ylim=(-0.1, 1.25))
    plt.show()

np.random.seed(0)
X2 = sp.stats.dirichlet((1, 1, 1)).rvs(1000)
plot_triangle(X2, kind="hex")
````
