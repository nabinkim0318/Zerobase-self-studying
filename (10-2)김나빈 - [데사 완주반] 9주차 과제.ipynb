````python
# 여기에 코드를 작성해주세요
import numpy as np
from IPython.display import Audio, display
#a-e-a-b-c#-a-b
def single_tone(frequency, sampling_rate=44100, duration=0.7):
  t = np.linspace(0, duration, int(sampling_rate))
  y = np.sin(2 * np.pi * frequency * t)
  return y

notes = 'G,D,G,A,A#,A'.split(',')
freqs = 261.62 * 2**(np.arange(0, len(notes)) / 12.)
notes = list(zip(notes, freqs))

sample = np.hstack([single_tone(f) for f in freqs])
display(Audio(sample, rate=44100))
````
