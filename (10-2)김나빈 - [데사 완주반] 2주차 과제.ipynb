### 1.(Python2_15) 아래 요구사항에 맞게 코드를 작성해주세요. (30점)
새로운 사용자의 email을 받으면 데이터프레임에 추가하는 함수를 만들려 합니다
- 함수는 데이터프레임과 email을 받고 데이터프레임을 반환합니다 \
(email은 영문자 소문자, 숫자로만 이루어져 있습니다)
- id는 항상 가장 마지막 id보다 1 큰 문자열입니다
- user_name은 email 주소의 id에 해당합니다
- domain은 email 컬럼에서 메일의 도메인만 가져옵니다
- 'user_name'과 'domain'은 split()으로도, re로도 가져올 수 있습니다

**결과 예시**
```python
# 함수 실행

print(df_update)
```
```

1  62040703          ko@naver.com          ko  naver

4  62040706        park@gmail.com        park  gmail

```

##### <코드>

````python
# 문제 풀기 전에 실행해주세요
import pandas as pd

data = {
    "id" : ['62040702','62040703','62040704','62040705','62040706'],
    "email" : ['kim@gmail.com','ko@naver.com','joh@gmail.com','lee@naver.com','park@gmail.com'],
    "user_name" : ['kim','ko','joh','lee','park'],
    "domain" : ['gmail','naver','gmail','naver','gmail']
}
df = pd.DataFrame(data)
df
````

````python
# 여기에 코드를 작성하세요
def update_user_data(df, email):
  domain = email.split("@")[1].split(".")[0]
  user = email.split("@")[0].split(".")[0]
  new_id = int(df.tail(1)["id"]) + 1
  df.loc[len(df)] = [new_id, email, user, domain]
  return df

df_update = update_user_data(df, 'fastcamp22@gmail.com')
df_update
````

````python
# 실행 코드
df_update = update_user_data(df, 'fastcamp22@gmail.com')
df_update
````

### 3.(Python2_26) 아래 요구사항에 맞게 코드를 작성해주세요. (40점)
보스턴 주택 가격 데이터셋을 활용하여 집값을 예측하려 합니다.
- 주어진 raw_data를 활용해주세요
- 가구 당 평균 방 개수('RM')와 고속도로 접근성 지수('RAD')로 집값('MEDV')을 회귀분석해주세요
- 모델 평가는 MAE로 해주세요
- 학습된 모델로 RM : 5 , RAD : 3 인 주택의 가격을 예측해보세요.
- 수행 순서
    - 데이터 전처리1 : 독립변수와 종속변수 나누기
    - 데이터 전처리2 : 학습데이터와 테스트데이터 나누기
    - 선형회기 모델 학습

##### <코드>

````python
# 문제 풀기 전에 실행해주세요
from sklearn.datasets import load_boston


dataset = load_boston()
raw_data = pd.DataFrame(dataset.data, columns=dataset.feature_names)
raw_data['MEDV'] = dataset.target
raw_data.head()
````

````python
# 여기에 코드를 작성하세요
import numpy as np
import pandas as pd
import pickle

from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.datasets import load_boston

dataset = load_boston()
raw_data = pd.DataFrame(dataset.data, columns=dataset.feature_names)
raw_data['MEDV'] = dataset.target

# 데이터 전처리 (독립 변수, 종속 변수)
df_x = raw_data[["RM", "RAD"]]
df_y = raw_data[["MEDV"]]

# 데이터 전처리 (학습데이터, 테스트 데이터)
train_x, test_x, train_y, test_y = train_test_split(
    df_x, df_y, test_size = 0.3, random_state = 1)

# 데이터 분석
model = linear_model.LinearRegression()
model.fit(train_x, train_y)

#성능평가 
pred_y = model.predict(test_x)
mae = mean_absolute_error(test_y, pred_y).round(2)
# 셋째자리에서 반올림. 즉 둘째짜리까지 살리기

# 예측 함수
def make_df(RM, RAD):
    return pd.DataFrame({"RM": [RM], "RAD":[RAD]})
RM, RAD = 5, 3
predict = float(model.predict(make_df(RM, RAD)).flatten().round(2))

print('MAE :', mae)
print('Predict :', predict)
````
