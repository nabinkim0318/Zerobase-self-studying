## 3. (ìˆ˜í•™_33) ë²¡í„°ë¡œ í‘œí˜„ëœ ë°ì´í„° aì™€ b ì‚¬ì´ì˜ ìœ í´ë¦¬ë“œ ê±°ë¦¬ì™€ ì½”ì‚¬ì¸ ìœ ì‚¬ë„, ê·¸ë¦¬ê³  ì½”ì‚¬ì¸ ê±°ë¦¬ë¥¼ êµ¬í•˜ì„¸ìš”. (25ì )
- ì¡°ê±´ : printë¬¸ì„ ì´ìš©í•˜ì—¬ ì½”ì‚¬ì¸ ìœ ì‚¬ë„ cosine_similarityì™€ ì½”ì‚¬ì¸ ê±°ë¦¬ cosine_distanceë¥¼ ì¶œë ¥í•˜ì„¸ìš”.

$$\mathbf{a} = \begin{bmatrix} 4 \\ 0 \\ 2 \\ 7 \end{bmatrix}$$
&nbsp;
$$\mathbf{b} = \begin{bmatrix} 1 \\ -2 \\ 4 \\ 0 \end{bmatrix}$$

- íŒíŠ¸: ë²¡í„°ì˜ normì€ np.linalg.norm(), ë‘ ë²¡í„°ì˜ ë‚´ì ì€ np.dot()ë¡œ êµ¬í•˜ì„¸ìš”.

\
- ì£¼ì–´ì§„ ë²¡í„°ë¥¼ ì½”ë“œë¡œ ì‘ì„±í•œë‹¤. (5ì )
- ìœ í´ë¦¬ë“œ ê±°ë¦¬ë¥¼ êµ¬í•œë‹¤. (6ì )
- ì½”ì‚¬ì¸ ìœ ì‚¬ë„ë¥¼ êµ¬í•œë‹¤. (7ì )
- ì½”ì‚¬ì¸ ê±°ë¦¬ë¥¼ êµ¬í•œë‹¤. (6ì )

````
# ì—¬ê¸°ì— ì½”ë“œë¥¼ ì‘ì„±í•´ì£¼ì„¸ìš”.
import numpy as np
a = np.array([[4], [0], [2], [7]])
b = np.array([[1], [-2], [4], [0]])

# ë²¡í„°ë¥¼ 2ì°¨ì› ë°°ì—´ë¡œ í•´ë„ ë˜ë‚˜ìš”? 
euclidean_distance = np.linalg.norm(a - b)
cosine_similarity = ((np.dot(a.T, b))/ (np.linalg.norm(a)*np.linalg.norm(b))).flatten()
cosine_distance = 1 - cosine_similarity

print(f"ìœ í´ë¦¬ë“œ ê±°ë¦¬:", euclidean_distance)
print(f"ì½”ì‚¬ì¸ ìœ ì‚¬ë„:", cosine_similarity)
print(f"ì½”ì‚¬ì¸ ê±°ë¦¬:", cosine_distance)
````

- ì¡°ê±´ : printë¬¸ì„ ì´ìš©í•˜ì—¬ ë­í¬-2 í–‰ë ¬ m2ê³¼ ì‹¤ì œ ë­í¬ rankë¥¼ ì¶œë ¥í•˜ì„¸ìš”.

$$\mathbf{x_1} = \begin{bmatrix} 1 \\ 1 \end{bmatrix}, \mathbf{x_2} = \begin{bmatrix} 1 \\ -1 \end{bmatrix}$$

- íŒíŠ¸: ë‘ ë²¡í„°ì˜ ë‚´ì ì€ @, í–‰ë ¬ì˜ ë­í¬ ê³„ì‚°ì€ np.linalg.matrix_rank()ë¡œ êµ¬í•˜ì„¸ìš”.

````
# ì—¬ê¸°ì— ì½”ë“œë¥¼ ì‘ì„±í•´ì£¼ì„¸ìš”.
import numpy as np
x1 = np.array([[1], [1]])
x2 = np.array(([1], [-1]))

m2 = x1@x1.T + x2@x2.T
rank = np.linalg.matrix_rank(m2)

print(m2)
print(rank)
````
