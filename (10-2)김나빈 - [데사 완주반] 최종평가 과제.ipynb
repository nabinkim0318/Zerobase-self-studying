## 4. (ÏàòÌïô_34) Î≤°ÌÑ∞ùêöÏùò Î≤°ÌÑ∞ùêõÏóê ÎåÄÌïú Ìà¨ÏòÅÏÑ±Î∂Ñ(projection)Í≥º ÏßÅÍµêÏÑ±Î∂Ñ(rejection)ÏùÑ Íµ¨ÌïòÏÑ∏Ïöî. (10Ï†ê)
- Ï°∞Í±¥: printÎ¨∏ÏùÑ Ïù¥Ïö©ÌïòÏó¨ Ìà¨ÏòÅÏÑ±Î∂Ñ projectionÍ≥º ÏßÅÍµêÏÑ±Î∂Ñ rejectionÏùÑ Ï∂úÎ†•ÌïòÏÑ∏Ïöî.

$$\mathbf{a} = \begin{bmatrix} 1 \\ 2 \end{bmatrix}$$
&nbsp;
$$\mathbf{b} = \begin{bmatrix} 2 \\ 0 \end{bmatrix}$$

- ÌûåÌä∏: Î≤°ÌÑ∞Ïùò normÏùÄ np.linalg.norm(), Îëê Î≤°ÌÑ∞Ïùò ÎÇ¥Ï†ÅÏùÄ np.dot()Î°ú Íµ¨ÌïòÏÑ∏Ïöî.

\
- Ìà¨ÏòÅÏÑ±Î∂ÑÏùò Í∏∏Ïù¥Î•º Íµ¨ÌïúÎã§. 
- Ìà¨ÏòÅÏÑ±Î∂Ñ Î≤°ÌÑ∞Î•º Íµ¨ÌïúÎã§. 
- ÏßÅÍµêÏÑ±Î∂Ñ Î≤°ÌÑ∞Î•º Íµ¨ÌïúÎã§. 

````
# Ïó¨Í∏∞Ïóê ÏΩîÎìúÎ•º ÏûëÏÑ±Ìï¥Ï£ºÏÑ∏Ïöî.
import numpy as np
a = np.array(([1], [2]))
b = np.array(([2], [0]))
a2 = np.squeeze(np.asarray(a))
b2 = np.squeeze(np.asarray(b))

#Ìà¨ÏòÅ ÏÑ±Î∂Ñ Í∏∏Ïù¥
dist_proj = np.dot(a2,b2) / np.linalg.norm(b)
#Ìà¨ÏòÅ ÏÑ±Î∂Ñ Î≤°ÌÑ∞
projection = dist_proj * (b / np.linalg.norm(b))
# ÏßÅÍµê ÏÑ±Î∂Ñ Î≤°ÌÑ∞
rejection = a - projection 

print(projection)
print(rejection)
````

## 5. (ÏàòÌïô_54) NumpyÎ•º ÏÇ¨Ïö©ÌïòÏó¨ Îã§Ïùå ÌñâÎ†¨ÏùÑ ÌäπÏûáÍ∞í Î∂ÑÌï¥ÌïòÏÑ∏Ïöî. ÎòêÌïú Îã§Ïãú Í≥±Ìï¥ÏÑú ÏõêÎûòÏùò ÌñâÎ†¨Ïù¥ ÎÇòÏò§Îäî Í≤ÉÏùÑ ÌôïÏù∏ÌïòÏÑ∏Ïöî. (10Ï†ê)

$$\mathbf{B} = \begin{bmatrix} 2 & 4 \\ 1 & 3 \\ 0 & 0 \\ 0 & 0 \end{bmatrix}$$

- Î¨∏Ï†ú 1: Ï∂ïÏÜåÌòïÏù¥ ÏïÑÎãå Î∞©Î≤ïÏúºÎ°ú Íµ¨ÌïòÏÑ∏Ïöî.
- Î¨∏Ï†ú 2: Ï∂ïÏÜåÌòï Î∞©Î≤ïÏúºÎ°ú Íµ¨ÌïòÏÑ∏Ïöî.
  - ÌûåÌä∏: ÌäπÏù¥Î∂ÑÌï¥Îäî svd()Î°ú Íµ¨ÌïòÏÑ∏Ïöî.

\
- Ï£ºÏñ¥ÏßÑ ÌñâÎ†¨ÏùÑ ÏΩîÎìúÎ°ú ÏûëÏÑ±ÌïúÎã§. 
- Ï£ºÏñ¥ÏßÑ ÌñâÎ†¨ÏùÑ ÌäπÏûáÍ∞í Î∂ÑÌï¥ÌïúÎã§. 
- ÌäπÏù¥Î∂ÑÌï¥Ìïú ÌñâÎ†¨ÏùÑ Í≥±Ìïú Í≤ÉÍ≥º ÏõêÎûòÏùò ÌñâÎ†¨Í≥º ÎèôÏùºÌï®ÏùÑ ÌôïÏù∏ÌïúÎã§. 

````
# Ïó¨Í∏∞Ïóê ÏΩîÎìúÎ•º ÏûëÏÑ±Ìï¥Ï£ºÏÑ∏Ïöî
import numpy as np
from numpy.linalg import svd

B = np.array([[2,4], [1,3], [0,0], [0,0]])
#Ï∂ïÏÜåÌòïÏù¥ ÏïÑÎãå Î∞©Î≤ï
U, S, VT = svd(B, full_matrices=True)
Sig = np.zeros((B.shape[0], B.shape[1]))
Sig[:B.shape[1], :B.shape[1]] = np.diag(S)
U @ Sig @ VT

#Ï∂ïÏÜåÌòï 
U2, S2, VT2 = svd(B, full_matrices=False)
U2 @ np.diag(S2) @ VT2
````
