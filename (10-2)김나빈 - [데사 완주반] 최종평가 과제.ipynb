## 4. (수학_34) 벡터𝐚의 벡터𝐛에 대한 투영성분(projection)과 직교성분(rejection)을 구하세요. (10점)
- 조건: print문을 이용하여 투영성분 projection과 직교성분 rejection을 출력하세요.

$$\mathbf{a} = \begin{bmatrix} 1 \\ 2 \end{bmatrix}$$
&nbsp;
$$\mathbf{b} = \begin{bmatrix} 2 \\ 0 \end{bmatrix}$$

- 힌트: 벡터의 norm은 np.linalg.norm(), 두 벡터의 내적은 np.dot()로 구하세요.

\
- 투영성분의 길이를 구한다. 
- 투영성분 벡터를 구한다. 
- 직교성분 벡터를 구한다. 

````
# 여기에 코드를 작성해주세요.
import numpy as np
a = np.array(([1], [2]))
b = np.array(([2], [0]))
a2 = np.squeeze(np.asarray(a))
b2 = np.squeeze(np.asarray(b))

#투영 성분 길이
dist_proj = np.dot(a2,b2) / np.linalg.norm(b)
#투영 성분 벡터
projection = dist_proj * (b / np.linalg.norm(b))
# 직교 성분 벡터
rejection = a - projection 

print(projection)
print(rejection)
````

$$\mathbf{B} = \begin{bmatrix} 2 & 4 \\ 1 & 3 \\ 0 & 0 \\ 0 & 0 \end{bmatrix}$$

- 문제 1: 축소형이 아닌 방법으로 구하세요.
- 문제 2: 축소형 방법으로 구하세요.
  - 힌트: 특이분해는 svd()로 구하세요.

\
- 주어진 행렬을 코드로 작성한다. 
- 주어진 행렬을 특잇값 분해한다. 
- 특이분해한 행렬을 곱한 것과 원래의 행렬과 동일함을 확인한다. 

````
# 여기에 코드를 작성해주세요
import numpy as np
from numpy.linalg import svd

B = np.array([[2,4], [1,3], [0,0], [0,0]])
#축소형이 아닌 방법
U, S, VT = svd(B, full_matrices=True)
Sig = np.zeros((B.shape[0], B.shape[1]))
Sig[:B.shape[1], :B.shape[1]] = np.diag(S)
U @ Sig @ VT

#축소형 
U2, S2, VT2 = svd(B, full_matrices=False)
U2 @ np.diag(S2) @ VT2
````

## 6. (수학_133) 몬티 홀 문제([Monty Hall Problem](https://github.com/pgmpy/pgmpy/blob/dev/examples/Monty%20Hall%20Problem.ipynb))를 pgmpy를 사용하여 구현하세요. (10점)
    세 개의 문 중에 하나를 선택하여 문 뒤에 있는 선물을 가질 수 있는 게임쇼가 있다. 
    한 문 뒤에는 자동차가 있고, 나머지 두 문 뒤에는 염소 두 마리가 각각 있다. 
    예를 들어, 참가자가 1번 문을 선택하면
    게임쇼 진행자는 남은 두 개의 문 중에 하나인 3번 문을 열어 문 뒤에 염소가 있음을 보여주면서
    1번 대신 2번 문으로 선택을 바꾸겠냐고 묻는다.
    이때 참가자는 선택을 바꾸는 것이 유리할까?

- 0, 1, 2는 문의 위치
- X: 참가자가 선택한 문을 나타내는 확률변수
- C: 자동차가 있는 문을 나타내는 확률변수
- H: 진행자가 열어준 문을 나타내는 확률변수

\
- 힌트 1: 자동차를 놓는 진행자는 참가자의 선택을 예측할 수 없고, 참가자는 자동차를 볼 수 없으므로 자동차의 위치와 참가자의 선택은 서로 독립적이다.
- 힌트 2: 진행자가 어떤 문을 여는지는 자동차의 위치 및 참가자의 선택에 좌우된다. 

  - 예를 들어, 자동차가 0번 문 뒤에 있고 참가자가 1번 문을 선택하면 진행자는 2번 문을 열 수 밖에 없다. 
  - 또한, 자동차가 1번 문 뒤에 있는데 참가자가 1번 문을 선택한 경우에 진행자는 0번 문과 2번 문 모두 열어도 되기 때문에 둘의 확률은 0.5로 동일하다.

\
- 주어진 데이터를 코드로 구현한다. 

````
# 여기에 코드를 작성해주세요.
!pip install pgmpy
from pgmpy.models import BayesianModel
from pgmpy.factors.discrete import TabularCPD

model = BayesianModel([('X', 'H'), ('C', 'H')])
cpd_X = TabularCPD('X', 3, [[0.33], [0.33], [0.33]])
cpd_C = TabularCPD('C', 3, [[0.33], [0.33], [0.33]])
cpd_H = TabularCPD('H', 3, [[0, 0, 0, 0, 0.5, 1, 0, 1, 0.5],
                            [0.5, 0, 1, 0, 0, 0, 1, 0, 0.5],
                            [0.5, 1, 0, 1, 0.5, 0, 0, 0, 0]],
                   evidence=['X', 'C'], evidence_card=[3, 3])
model.add_cpds(cpd_X, cpd_C, cpd_H)
model.check_model()
````

````
from pgmpy.inference import VariableElimination
inference = VariableElimination(model)
posterior = inference.query(['C'], evidence={'X': 1, 'H': 2}, show_progress=False)
print(posterior)
````
