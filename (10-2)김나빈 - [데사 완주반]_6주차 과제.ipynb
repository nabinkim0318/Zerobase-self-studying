- 인구가 500만이 넘는 도시를 조회합니다
- 조회한 도시의 대륙을 조회합니다
- GROUP BY Continent
- 한 국가에 여러 도시가 있기 때문에 중복제거를 하면 안됩니다.

##### <코드>

- 여기에 쿼리를 작성하세요
```sql
USE world;
SELECT country.Continent, count(city.name)
FROM 
	(SELECT countrycode, name, population
	FROM city
	WHERE population > 5000000) as city
JOIN 
	(SELECT code, continent
	FROM country) AS country
ON city.countrycode = country.code
GROUP BY Continent
```

##### <힌트>

- 관계형 데이터베이스에서 원하는 데이터를 조회할때 ERD를 살펴보는 것이 도움이 됩니다


##### <코드>

- 여기에 쿼리를 작성하세요
```sql
USE sakila;
SELECT actor.name, count(actor.actor_id) as Count
FROM
	(SELECT concat(first_name, " ", last_name) as name, actor_id
	FROM actor) As actor
JOIN
	(SELECT actor_id FROM film_actor) As film_actor
ON actor.actor_id = film_actor.actor_id
GROUP BY film_actor.actor_id
ORDER BY Count DESC
LIMIT  10
```
